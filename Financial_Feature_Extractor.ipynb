SCRAPING THE DATA

```python
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

# Function to create a directory if it doesn't exist
def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Function to convert a value to float if possible, else return the original value
def convert_to_float(value):
    try:
        return float(value.replace(',', '')) if isinstance(value, str) else float(value)
    except ValueError:
        return value

# Create a directory to store all financial statement folders
main_directory = os.path.join(os.getcwd(), "Financial_Statements")
create_directory(main_directory)

# Read the CSV file
file_path = "Moneycontrol Financial Statement Links.csv"
data = pd.read_csv(file_path)

# Loop through each row in the CSV
for index, row in data.iterrows():
    company_name = row["Company Name"]

    # Create a directory for the company within Financial_Statements
    company_directory = os.path.join(main_directory, company_name)
    create_directory(company_directory)

    # Function to fetch data and save in respective folders
    def fetch_and_save_data(url, folder_name):
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        table = soup.find("table", class_="mctable1")
        if table:
            file_name = os.path.join(company_directory, f"{folder_name}.csv")
            with open(file_name, "w", newline="") as csvfile:
                csv_writer = csv.writer(csvfile)
                for row in table.find_all("tr"):
                    csv_writer.writerow([convert_to_float(cell.text.strip()) for cell in row.find_all("td")])

    # Fetch and save data for each URL
    for column_name in ["Moneycontrol_BS1", "Moneycontrol_BS2", "Moneycontrol_PL1", "Moneycontrol_PL2",
                        "Moneycontrol_CF1", "Moneycontrol_CF2", "Moneycontrol_Ratio1", "Moneycontrol_Ratio2"]:
        url = row[column_name]
        if url:
            sheet_name = column_name.split()[-1]  # Extracting the sheet name from the column name
            fetch_and_save_data(url, f"{sheet_name}")


    print(index, end =" ")

print("Data retrieval and storage completed.")
```

REPLACING " " WITH "_"

```python
import os
import csv

directory = "/home/student/Desktop/Dataset Creation/Statements to Dataset/Financial_Statements"

def replace_spaces_with_underscore(file_path):
    print(f"Processing file: {file_path}")
    with open(file_path, 'r') as file:
        csv_reader = csv.reader(file)
        rows = list(csv_reader)
        if rows:
            for row in rows:
                if row[0].strip().count(' ') > 0:
                    row[0] = row[0].replace(' ', '_')

    with open(file_path, 'w', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerows(rows)
        print("Changes made successfully!")

for root, dirs, files in os.walk(directory):
    for file_name in files:
        if file_name.endswith('.csv'):
            file_path = os.path.join(root, file_name)
            replace_spaces_with_underscore(file_path)
```
