```python
import os
import pandas as pd
main_directory = "/home/student/Desktop/Dataset Creation/Statements to Dataset/Financial_Statements"

folders = [folder for folder in os.listdir(main_directory) if os.path.isdir(os.path.join(main_directory, folder))]

count = 0

# Loop through each folder
for folder_name in folders:
    print(f"Processing folder: {folder_name}")
    folder_path = os.path.join(main_directory, folder_name)

    bs1 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_BS1.csv")).filter(regex='^(?!Unnamed)')
    bs2 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_BS2.csv")).filter(regex='^(?!Unnamed)')
    merged_bs = pd.concat([bs1, bs2.iloc[:, 1:]], axis=1)
    merged_bs.to_csv(os.path.join(folder_path, "Merged_Balance_Sheet.csv"), index=False)

    pl1 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_PL1.csv")).filter(regex='^(?!Unnamed)')
    pl2 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_PL2.csv")).filter(regex='^(?!Unnamed)')
    merged_pl = pd.concat([pl1, pl2.iloc[:, 1:]], axis=1) 
    merged_pl.to_csv(os.path.join(folder_path, "Merged_Profit_Loss.csv"), index=False)

    cf1 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_CF1.csv")).filter(regex='^(?!Unnamed)')
    cf2 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_CF2.csv")).filter(regex='^(?!Unnamed)')
    merged_cf = pd.concat([cf1, cf2.iloc[:, 1:]], axis=1)  
    merged_cf.to_csv(os.path.join(folder_path, "Merged_Cash_Flow.csv"), index=False)

    ratio1 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_Ratio1.csv")).filter(regex='^(?!Unnamed)')
    ratio2 = pd.read_csv(os.path.join(folder_path, "Moneycontrol_Ratio2.csv")).filter(regex='^(?!Unnamed)')
    merged_ratio = pd.concat([ratio1, ratio2.iloc[:, 1:]], axis=1) 
    merged_ratio.to_csv(os.path.join(folder_path, "Merged_Ratio.csv"), index=False)
    
    print(count, end = " ")
    count = count+1
    

print("Processing completed for all folders.")
```
